```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## *Analisis de Facturación*

```python
total_tienda1 = tienda['Precio'].sum()
total_tienda2 = tienda2['Precio'].sum()
total_tienda3 = tienda3['Precio'].sum()
total_tienda4 = tienda4['Precio'].sum()

print("Total tienda 1:", total_tienda1)
print("Total tienda 2:", total_tienda2)
print("Total tienda 3:", total_tienda3)
print("Total tienda 4:", total_tienda4)
```

## *Ventas por Categoria*

```python
tienda.columns
```

```python
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for nombre, df in tiendas.items():
    print("\n", nombre)
    print(df.groupby('Categoría del Producto')
            .size()
            .sort_values(ascending=False))
```

## *Valoración Media por tienda*

```python
print("Categoría más vendida - Tienda 1:", tienda['Categoría del Producto'].value_counts().idxmax())
print("Categoría más vendida - Tienda 2:", tienda2['Categoría del Producto'].value_counts().idxmax())
print("Categoría más vendida - Tienda 3:", tienda3['Categoría del Producto'].value_counts().idxmax())
print("Categoría más vendida - Tienda 4:", tienda4['Categoría del Producto'].value_counts().idxmax())
```

## *Productos más y menos vendidos*

```python
print("Tienda 1")
print("Más vendido:")
print(tienda['Producto'].value_counts().head(1))
print("Menos vendido:")
print(tienda['Producto'].value_counts().tail(1))

print("\nTienda 2")
print("Más vendido:")
print(tienda2['Producto'].value_counts().head(1))
print("Menos vendido:")
print(tienda2['Producto'].value_counts().tail(1))

print("\nTienda 3")
print("Más vendido:")
print(tienda3['Producto'].value_counts().head(1))
print("Menos vendido:")
print(tienda3['Producto'].value_counts().tail(1))

print("\nTienda 4")
print("Más vendido:")
print(tienda4['Producto'].value_counts().head(1))
print("Menos vendido:")
print(tienda4['Producto'].value_counts().tail(1))
```

## *Valor del envio promedio por tienda*

```python
print("Promedio envío - Tienda 1:", tienda['Costo de envío'].mean())
print("Promedio envío - Tienda 2:", tienda2['Costo de envío'].mean())
print("Promedio envío - Tienda 3:", tienda3['Costo de envío'].mean())
print("Promedio envío - Tienda 4:", tienda4['Costo de envío'].mean())
```

## *Visualización*

```python
import matplotlib.pyplot as plt
```

```python
colores = {
    "Tienda 1": "#4E79A7",
    "Tienda 2": "#F28E2B",
    "Tienda 3": "#59A14F",
    "Tienda 4": "#E15759"
}

tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}
```

```python
nombres = []
ingresos = []
lista_colores = []

for nombre, df in tiendas.items():
    nombres.append(nombre)
    ingresos.append(df["Precio"].sum())
    lista_colores.append(colores[nombre])

plt.figure(figsize=(8,5))
plt.bar(nombres, ingresos, color=lista_colores)

plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos Totales")

plt.show()
```

```python
nombres = []
calificaciones = []

for nombre, df in tiendas.items():
    nombres.append(nombre)
    calificaciones.append(df["Calificación"].mean())

plt.figure(figsize=(8,5))

for i in range(len(nombres)):
    plt.scatter(nombres[i], calificaciones[i],
                color=colores[nombres[i]],
                s=100)

plt.plot(nombres, calificaciones, color="gray")

plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")

plt.show()
```

```python
top5 = tienda['Producto'].value_counts().head(5)

plt.figure(figsize=(8,5))
plt.bar(top5.index, top5.values, color=colores["Tienda 1"])

plt.title("Top 5 Productos - Tienda 1")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

colores = {
    "Tienda 1": "#4E79A7",
    "Tienda 2": "#F28E2B",
    "Tienda 3": "#59A14F",
    "Tienda 4": "#E15759"
}

tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

plt.figure(figsize=(8,6))

for nombre, df in tiendas.items():
    plt.scatter(df["lon"], df["lat"],
                alpha=0.3,
                color=colores[nombre],
                label=nombre)

plt.title("Distribución Geográfica de Ventas")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()

plt.show()
```
